1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r', encoding='latin-1')
print(type(archivo))
```

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}

faltantes = []
emisiones_totales = 0

lineas = archivo.readlines()
for linea in lineas:
    campos = linea.rstrip().split('|')
    dicc_emisiones['cod_pais'].append(campos[0])
    dicc_emisiones['nom_pais'].append(campos[1])
    dicc_emisiones['region'].append(campos[2])
    dicc_emisiones['anio'].append(campos[3])
    dicc_emisiones['co2'].append(campos[4])
    dicc_emisiones['co2_percapita'].append(campos[5])

    if campos[2] == 'América Latina y Caribe' and campos[3] == '2010' and campos[4] != '':
        emisiones_totales += float(campos[4].replace('.','').replace(',','.'))
        

#print(dicc_emisiones)

#Elementos faltantes
print('cod_pais', dicc_emisiones['cod_pais'].count(''))
print('nom_pais', dicc_emisiones['nom_pais'].count(''))
print('region', dicc_emisiones['region'].count(''))
print('anio', dicc_emisiones['anio'].count(''))
print('co2', dicc_emisiones['co2'].count(''))
print('co2_percapita', dicc_emisiones['co2_percapita'].count(''))

#total emisiones latam y caribe
print('\nemisiones totales latam y caribe', emisiones_totales)
```
